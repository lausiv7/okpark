```python
def indicator_constraints_add(problem,k):
# add the indicator constraints
    ic_dict = {}

    for i in range(0,22-k):

        ic_dict["lin_expr"]     = cplex.SparsePair(ind = ["c"+str(i),"c"+str(i+1)],
                                                   val = [1.0,-1.0])
        ic_dict["rhs"]          = 0.0
        ic_dict["sense"]        = "E"
        ic_dict["indvar"]       = "x"+str(i)
        ic_dict["complemented"] = 1
        problem.indicator_constraints.add(**ic_dict)

    for i in range(0,22-k):

        ic_dict["lin_expr"]     = cplex.SparsePair(ind = ["c"+str(i),"c"+str(i+1)],
                                                   val = [1.0,1.0])
        ic_dict["rhs"]          = 1.0
        ic_dict["sense"]        = "E"
        ic_dict["indvar"]       = "x"+str(i)
        ic_dict["complemented"] = 0
        problem.indicator_constraints.add(**ic_dict)
```
